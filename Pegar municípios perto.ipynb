```python
import pandas as pd
import requests
import json
import csv

def isnan(value):
  try:
      import math
      return not(math.isnan(float(value)))
  except:
      return True

excel = pd.read_excel (r'Lista Municpios, solo, data de plantio - INMET - Proagro.xlsx', 'Municpios')
df = pd.DataFrame(excel, columns= excel)

row_list = []
row_list.append(["centroideY","centroideX","nome","URL","codigoStr"])
with open("municipiosPerto.csv", 'w', newline='') as file4:
    writer4 = csv.writer(file4)
    writer4.writerows(row_list)
i = 0;
for colunas in df:
    try:
        i=i+1
        print(i)
        url = colunas[3]
        response = requests.get(url,headers={"Content-Type": "application/json;charset=UTF-8"})
        parsed_json = (json.loads(response.text))
        
        print(parsed_json['estacao']['codigoStr'])
        row_list.append([colunas[0],colunas[1],colunas[2],colunas[3],parsed_json['estacao']['codigoStr']])
    except:
        continue
    with open('municipiosPerto.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(row_list)
    row_list = []
                    
```
